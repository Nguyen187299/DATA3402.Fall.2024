3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
class Matrix:
    def __init__(self, n, m, data=None):
        self.n = n
        self.m = m
        if data is None:
            self.data=[[0.0 for i in range(m)] for i in range(n)]
        else:
            self.data = data

    def __getitem__(self, key):
        if isinstance(key, tuple):
            i, j = key
            return self.data[i][j]
        else:
            return self.data[key]

    def __setitem__(self, key, value):
        if isinstance(key, tuple):
            i, j = key
            self.data[i][j] = value
        else:
            self.data[key] = value

    def __str__(self):
        return '\n'.join([' '.join(f"{int(item)}" for item in row) for row in self.data])

def constant(n, m, c):
    return Matrix(n, m, [[float(c) for i in range(m)] for i in range(n)])

def zeros(n, m):
    return Matrix(n, m)

def ones(n, m):
    return Matrix(n, m, [[1.0 for i in range(m)] for i in range(n)])

def eye(n):
    identity_matrix = Matrix(n, n)
    for i in range(n):
        identity_matrix[i, i]=1 
    return identity_matrix

print("Constant matrix")
print(constant(3, 3, 5))
print("Zeros matrix")
print(zeros(2, 3))
print("Ones matrix")
print(ones(2, 3))
print("Identity matrix")
print(eye(3))
```
